### Задание 1. Написать функцию на Python
Напиши функцию, которая удалит дубликаты в списке. Список не отсортирован. Необходимо сохранить порядок сортировки оригинального списка.

```python
examples = [[1, 2, 3, 1], 
            [1, 3, 2, 1, 5, 3, 5, 1, 4], 
            [1, 1, 1, 1], 
            []]
answers  = [[1, 2, 3], 
            [1, 3, 2, 5, 4], 
            [1], 
            []]
```

```python
def delete_copies(origin_list):
    unique_items = set()
    new_list = []
    for item in origin_list:
        if item not in unique_items:
            unique_items.add(item)
            new_list.append(item)
    return new_list
```

Асимптотическая сложность функции - О(n)

```python
for i in range(len(examples)):
    assert(delete_copies(examples[i]) == answers[i])
```

### Задание 3. Решить задачу
Подкинули монету N раз. Кол-во случаев, когда выпал орёл, на 10% больше, чем кол-во случаев, когда выпала решка. При каком N мы можем сказать, что монета «нечестная» (орёл и решка выпадают с разной вероятностью)?

### Решение
У нас есть выборка из N бросков монетки: причем в 0.55N из них выпал орел, и в остальных 0.45N выпала решка.

Вследствие центральной предельной теоремы мы знаем, что все выборочные средние имеют нормальное распределение вокруг среднего генеральной совокупности. Т.е. если мы проводим много разных экспериментов с монеткой (например, будем считать, что 0 - это выпадение решки, а 1 - орла), в каждом из них будет какое-то свое значение среднего (в задаче это среднее = 0.55). Но совокупность средних значений экспериментов будет выглядеть как нормальное распределение со средним 0.5.

Добавим 2 гипотезы:
    
    H0 - среднее == 0.5
    H1 - среднее != 0.5

Для начала мы примем гипотезу Н0.

Стандартная ошибка среднего $se$ вычисляется как: 
$$se = \frac{\sigma}{\sqrt{N}}$$

$\sigma$ - среднеквадратичное отклонение

Чтобы посчитать (в кол-ве $\sigma$) отклонение нашего выборочного среднего, необходимо вычислить:

$$z = \frac{\hat{x} - ex}{se}$$

$\hat{x}$ - выборочное среднее, $ex$ - среднее генеральной совокупности

$$z = \frac{0.55 - 0.5}{se} = \frac{0.05 * \sqrt{N}}{\sigma}$$

Далее нам интересно узнать, какова вероятность получить такое или большее отклонение (закрашенная площадь под графиком)



Если площадь под графиком будет < 0.05 (p-уровень), то можно будет принять альтернативную гипотезу Н1.

Площадь под графиком < 0.05, если отклонение больше, чем 2$\sigma$

$$ 0.05 * \sqrt{N} > 2 $$

$$ \sqrt{N} > \frac{2}{0.05} $$

$$ N > 1600 $$

## Ответ:
Монетку можно считать нечестной при количестве бросков N, большем 1600.
